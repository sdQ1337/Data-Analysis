```python
import pandas as pd
file_path = "商品销售数据.xlsx"
info_df = pd.read_excel(file_path, sheet_name="信息表")
info_df
```

```python
sales_df = pd.read_excel(file_path, sheet_name="销售数据表")
sales_df
```

```python
merged_df = pd.merge(info_df, sales_df, on="商品编号", how="right")
merged_df
```

```python
merged_df['销售金额'] = merged_df['商品销售价'] * merged_df['订单数量']
merged_df
```

```python
merged_df['月份'] = pd.to_datetime(merged_df['订单日期']).dt.month
merged_df
```

```python
monthly_sales = merged_df.groupby('月份')['销售金额'].sum().reset_index()
monthly_sales
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['月份'], monthly_sales['销售金额'], marker='o', linestyle='-', color='b')
plt.title('商品每月销售金额折线图', fontsize=14)
plt.xlabel('月份', fontsize=12)
plt.ylabel('销售金额（元）', fontsize=12)
plt.xticks(monthly_sales['月份'])
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
category_monthly_sales = merged_df.groupby(['月份', '商品大类'])['销售金额'].sum().unstack()
category_monthly_sales
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
category_monthly_sales.plot(kind='bar', width=0.8)
plt.title('各商品大类月度销售额对比', fontsize=14)
plt.xlabel('月份', fontsize=12)
plt.ylabel('销售金额（元）', fontsize=12)
plt.xticks(rotation=0)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.legend(title='商品大类', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7, axis='y')
plt.tight_layout()
plt.show()
```
